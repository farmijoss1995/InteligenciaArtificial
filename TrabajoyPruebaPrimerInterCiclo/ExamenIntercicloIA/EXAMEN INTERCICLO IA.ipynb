```python
class Alcalde(StructuredNode):
    nombre = StringProperty(unique_index=True)

class Tweets(StructuredNode):
    id_Tts = UniqueIdProperty()
    Noticias = StringProperty(unique_index=False)
    Date = StringProperty(unique_index=False)
    Retweets = StringProperty(unique_index=False)
    usuario = RelationshipTo('Usuario', 'tweetUsuario')
    alcalde = RelationshipTo('Alcalde', 'tweetAlcalde')
    
class Usuario(StructuredNode):
    idUsuario = UniqueIdProperty()
    nombre = StringProperty(unique_index=True)
    tweet = RelationshipFrom('Tweets','tweetUsuario')
    
    
```
